- Their are some problematic values in all these columns so we need to clean them first before typechange 

```python
def clean_damage(val):
    if pd.isnull(val):
        return np.nan
    
    val = str(val).replace(',', '').strip()
    
    # Case 1: Range (e.g., 30-350)
    if '-' in val:
        parts = re.split(r'-', val)
        if len(parts) == 2:
            return (float(parts[0]) + float(parts[1])) / 2
    
    # Case 2: Fraction (e.g., 106/50)
    if '/' in val:
        parts = re.split(r'/', val)
        if len(parts) == 2:
            return float(parts[0]) / float(parts[1])
    
    # Case 3: Normal number
    try:
        return float(val)
    except:
        return np.nan
```

```python
df['damage'] = df['damage'].apply(clean_damage)
df['dps'] = df['dps'].apply(clean_damage)
df['hit_speed'] = df['hit_speed'].apply(clean_damage)
df["range"]=df["range"].apply(clean_damage)
```

```python
# Extracted the values of spawn unit level and original card level from card_level. 
df[['card_level_f', 'spawn_level']] = df['card_level'].str.extract(
    r'(\d+)\s*\((\d+)\)'
)
```

```python
def split_hp(val):
    if pd.isnull(val):
        return pd.Series([np.nan, np.nan])
    
    val = str(val).replace(',', '').strip()
    
    # Case 1: Shield format (e.g., 696 (+696))
    match = re.match(r'(\d+)\s*\(\+(\d+)\)', val)
    if match:
        base = float(match.group(1))
        shield = float(match.group(2))
        return pd.Series([base, shield])
    
    # Case 2: Fraction-like (e.g., 169/110) → treat as base/shield
    if '/' in val:
        parts = val.split('/')
        if len(parts) == 2:
            return pd.Series([float(parts[0]), float(parts[1])])
    
    # Case 3: Normal value → only base HP
    try:
        return pd.Series([float(val), 0])  # no shield
    except:
        return pd.Series([np.nan, np.nan])
```

```python
# values are allocated to shield and base from original column and then it droppep
df[['base_hp', 'shield_hp']] = df['sheild'].apply(split_hp)
df.drop('sheild', axis=1, inplace=True)
```
